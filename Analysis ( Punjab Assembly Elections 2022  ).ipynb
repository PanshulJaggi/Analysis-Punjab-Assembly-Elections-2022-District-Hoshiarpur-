```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns

# load the data of Punjab Assembly Elections 2022.
data = pd.read_csv("C:\\Users\\pansh\\Downloads\\Punjab Assembly Elections 2022.csv")
data
```

```python
# how many Winning Candidates in Punjab are of each type (SC or GEN).
data['Type'].value_counts()
```

```python
# the top 5 most frequently occurring districts in the Punjab.
data['District'].value_counts()[0:5]
```

```python
# Data Visualization of the top 5 most frequently occurring districts in the Punjab.
a = data['District'].value_counts()[0:5]
plt.bar(list(a.keys()),list(a),color='red',edgecolor='k')

plt.title('The top 5 most frequently occurring districts in the Punjab.')
plt.xlabel('District', fontdict={'fontsize': 15, 'fontname': 'Times New Roman'},color='k')
plt.ylabel('No. of Districts', fontdict={'fontsize': 15, 'fontname': 'Times New Roman'},color='k')
plt.show()
```

```python
# Find the number of assembly seats won by each political party in the Punjab Assembly Elections 2022. 
data['Party'].value_counts()
```

```python
# Data Visulationzation of the number of assembly seats won by each political party in the Punjab Assembly Elections 2022. 

a = data['Party'].value_counts()
plt.figure(figsize=(13,6))
aa = sns.barplot(x=a.index,y=a.values,width=0.2)

plt.title('The number of assembly seats won by each political party in the Punjab Assembly Elections 2022.')

plt.xlabel('Party', fontdict={'fontsize': 15, 'fontname': 'Times New Roman'},color='k')
plt.ylabel('No. of Seats', fontdict={'fontsize': 15, 'fontname': 'Times New Roman'},color='k')
for i in aa.containers:
    aa.bar_label(i)
plt.show()
```

```python
# shows the data where the party that won in the election was Aam Aadmi Party.
data.loc[data['Party']=='Aam Aadmi Party']
```

```python
# find the name of the winning candidate for AAP who received the highest percentage of votes in the Punjab assembly elections 2022.
aap = data.loc[data['Party']=='Aam Aadmi Party']
aap.loc[aap['% of Votes'].idxmax(),['Winning Candidate','District']]
```

```python
# find the name of the winning candidate for AAP who received the lowest percentage of votes in the Punjab assembly elections 2022.
aap = data.loc[data['Party']=='Aam Aadmi Party']
aap.loc[aap['% of Votes'].idxmin(),['Winning Candidate','AC Name','District']]
```

```python
#  Analysis of the Punjab Assembly Election 2022's  data related to only the Hoshiarpur district in Punjab.
hoshiarpur = data.loc[data['District']=='Hoshiarpur']
hoshiarpur
```

```python
# how many Winning Candidates in Hoshiarpur are of each type (SC or GEN).
hoshiarpur['Type'].value_counts()
```

```python
# get the name of the winning candidate with the highest number of EVM votes in the Hoshiarpur district elections
hoshiarpur.loc[hoshiarpur['EVM Votes'].idxmax(), 'Winning Candidate']
```

```python
# get the name of the winning candidate with the lowest number of EVM votes in the Hoshiarpur district elections
hoshiarpur.loc[hoshiarpur['EVM Votes'].idxmin(), 'Winning Candidate']
```

```python
# calculates the maximum EVM votes received by each party in Hoshiarpur.
hoshiarpur.groupby(['Party'],as_index=True)['EVM Votes'].max()
```

```python
# calculates the maximum Total Votes\t votes received by each party in Hoshiarpur.
hoshiarpur.groupby(['Party'],as_index=True)['Total Votes\t'].max()
```

```python
# Data Visualization of the maximum Total Votes\t votes received by each party in Hoshiarpur.

a = hoshiarpur.groupby(['Party'], as_index=True)['Total Votes\t'].max()
plt.figure(figsize=(7,7))
aa = sns.barplot(x=a.index, y=a.values,edgecolor='k',width=0.4)
for i in aa.containers:
    aa.bar_label(i)

plt.title('the maximum Total Votes received by each party in Hoshiarpur.',fontdict={'fontsize': 14, 'fontname': 'Times New Roman'})
plt.xlabel('Party', fontdict={'fontsize': 14, 'fontname': 'Times New Roman'})
plt.ylabel('No. of votes', fontdict={'fontsize': 14, 'fontname': 'Times New Roman'})

plt.show()
```

```python
#  finds the maximum value of the '% of Votes' for each party in Hoshiarpur.
hoshiarpur.groupby(['Party'],as_index=True)['% of Votes'].max()
```

```python
# Data Visualization ofthe maximum value of the '% of Votes' for each party in Hoshiarpur.


a = hoshiarpur.groupby(['Party'], as_index=True)['% of Votes'].max()
plt.figure(figsize=(7,7))
aa = sns.barplot(x=a.index, y=a.values,edgecolor='k',width=0.4)
for i in aa.containers:
    aa.bar_label(i)

plt.title('the maximum Total Votes received by each party in Hoshiarpur.',fontdict={'fontsize': 14, 'fontname': 'Times New Roman'})
plt.xlabel('Party', fontdict={'fontsize': 14, 'fontname': 'Times New Roman'})
plt.ylabel('% of votes', fontdict={'fontsize': 14, 'fontname': 'Times New Roman'})

plt.show()
```

```python
# Data visualize of the percentage of votes received by the winning candidates of each party in the Hoshiarpur district of Punjab, India.
sns.set(style='white')
plt.figure(figsize=(18,10))
s = sns.barplot(x="Winning Candidate",y='% of Votes',hue='Party',data=hoshiarpur)
for j in s.containers:
    s.bar_label(j)

plt.title('Visualize of the percentage of votes received by the winning candidates of each party in the Hoshiarpur district of Punjab.')
plt.legend(title='Party', bbox_to_anchor=(1, 0, 0.1, 1), prop={'size': 20, 'family': 'Times New Roman'})
plt.xlabel('Winning Candidates', fontdict={'fontsize': 20, 'fontname': 'Times New Roman'})
plt.ylabel('% of votes', fontdict={'fontsize': 20, 'fontname': 'Times New Roman'})
plt.grid()
plt.show()
```
